## Implementing SOM using Minisom

```python
import numpy as np
import math 
import random
from minisom import MiniSom
import pandas as pd
```

```python
data = pd.read_csv('credit_card_normalized.csv')
```

```python
data = (data - np.mean(data, axis=0)) / np.std(data, axis=0)
data = data.values
data.shape[1]
print(data)
```

```python
som_shape = (3,2)
som = MiniSom(som_shape[0],som_shape[1],data.shape[1],
              sigma=0.5,learning_rate=0.03,
              neighborhood_function='gaussian',
              random_seed=42)

som.train_batch(data,100,verbose=True)
```

```python
# each neuron represents a cluster
winner_coordinates = np.array([som.winner(x) for x in data]).T

cluster_index = np.ravel_multi_index(winner_coordinates, som_shape)
```

### Clusters formed

```python
import matplotlib.pyplot as plt
%matplotlib inline

# plotting the clusters using the first 2 dimentions of the data
for c in np.unique(cluster_index):
    plt.scatter(data[cluster_index == c, 0],
                data[cluster_index == c, 1], label='cluster='+str(c), alpha=.5)

# plotting centroids
for centroid in som.get_weights():
    plt.scatter(centroid[:, 0], centroid[:, 1], marker='x', 
                s=80, linewidths=5, color='k', label='centroid')
plt.legend();
```
